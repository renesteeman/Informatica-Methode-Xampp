```python
# importeer de nodige packages
import numpy as np #wordt gebruikt voor datamanipulatie
import pandas as pd #wordt gebruikt voor datamanipulatie
import matplotlib.pyplot as plt #wordt gebruikt voor grafieken
import seaborn as sns #wordt gebruikt om de grafieken mooier te maken
import statsmodels.api as sm #wordt gebruikt voor berekingen

#Lees de data in, dit zijn de waardes die de grote van een huis tenopzichte de prijs ervan weergeven.
data = pd.read_csv('dataset.csv') #lees de dataset in die in deze map staat met de naam 'dataset.csv'

#stel de data voor de x- en y-as in
y = data['price'] #gebruik de prijzen voor de y-as
x = data['size'] #gebruik de grote van de huizen voor de x-as

#toon de data
plt.scatter(x,y) #maak een scatterplot (een soort grafiek)
plt.xlabel('Size',fontsize=20) #stel de tekst voor de x-as in
plt.ylabel('Price',fontsize=20) #stel de tekst voor de y-as in
plt.show() #toon de grafiek

#voor de machine learning uit
x_ML = sm.add_constant(x) #geef aan wat de x-as is voor de machine learning
results = sm.OLS(y,x_ML).fit() #train het model
results.summary() #geef een samenvatting van de resultaten

#Je weet dat een rechte lijn te noteren is als y=ax+b met a als stijlheid van de lijn en b de startwaarde.
#Je ziet dat de waarde voor de coëfficiënt van const 1.019e+05 is (tweede tabel en tweede rij), oftewel 101900. Deze 101900 is de waarde van b.
#Je ziet ook dat de waarde voor de coëfficiënt van size 223.1787 is, dit is de waarde van a.
#Je kunt nu de lijn y=223.1787x + 101900 opstellen.
```

```python
#laat de resultaten van de machine learning zien
plt.scatter(x,y) #stel de grafiek in
yhat = x*223.1787+101900 #toon de gevonden lijn 
fig = plt.plot(x,yhat, lw=4, c='orange', label ='regression line') #stel de weergave van de grafiek in
plt.xlabel('Size', fontsize = 20) #stel de tekst voor de x-as in
plt.ylabel('Price', fontsize = 20) #stel de tekst voor de y-as in
plt.show() #toon de grafiek
```
